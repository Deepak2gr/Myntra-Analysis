```python
orders["Month"] = orders["Date"].dt.strftime("%B")
orders
```

```python
orders["Total Price"] = orders["Original Price"] - (orders["Original Price"] * orders["Discount%"])
orders
```

#Joinning order and product table

```python
op = pd.merge(left = orders, right= product, on = "Product ID", how = "inner" )
op
#order and product table
```

#joing order and customers table

```python
oc = pd.merge(left = orders, right= customer, on = "Customer ID", how = "inner" )
oc
#order and customers table
```

```python
MbyP = orders.groupby("Month").agg({"Total Price":"sum"})
MbyP
# month by price
```

```python
cbyp= op.groupby("Category").agg({"Total Price":"sum"})
cbyp
#Category by price
```

```python
Bbyp= op.groupby("Brand Name").agg({"Total Price":"sum"}).head(10)
Bbyp
#Brand Name by price
```

#top 10 brand by the total price

```python
to_10_b = Bbyp.sort_values(by = "Total Price",ascending=False).head(10)
to_10_b
```

```python
cbyp= oc.groupby("City").agg({"Total Price":"sum"}).head(5)
cbyp
#city by price
```

```python
sbyp= oc.groupby("State").agg({"Total Price":"sum"})
sbyp
#state by price
```

```python
cabyp= oc.groupby("Customer Age").agg({"Total Price":"sum"})
cabyp
#customer age by price
```

```python
bins = [5,18,30,45,60,100]
labels=["0-18","19-30","31-45","45-60","60+"]
oc["Age Group"] = pd.cut(oc["Customer Age"] , bins=bins , labels=labels,right= True)
oc
```

```python
caGbyp= oc.groupby("Age Group").agg({"Total Price":"sum"})
caGbyp
#customer age Group by price
```

```python
sns.barplot(x = MbyP.index, y = MbyP["Total Price"], order = ["January", "February", "March",'April',"May","June","July","August","September","October","November", "December"],hue = MbyP.index, palette = "muted")
plt.xticks(rotation = 60)
plt.show()
```

```python
plt.pie(caGbyp["Total Price"], labels = caGbyp.index, autopct = "%1.2f%%")
plt.show()
```
